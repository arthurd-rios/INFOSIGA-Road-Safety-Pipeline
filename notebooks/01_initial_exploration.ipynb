```python
import pandas as pd
```

```python
# Leitura da base bruta completa do Infosiga (2022-2025)

df = pd.read_csv('../data_raw/sinistros_2022-2025.csv', sep=';', encoding='latin1', low_memory = False)
```

```python
# Filtragem das rodovias de interesse
# Foi selecionado um mix de rodovias (Boas/Médias/Ruins) para treinar o modelo com diferentes cenários de infraestrutura

mask = df['logradouro'].str.contains('SP 348|SP 280|SP 070|SP 330|SP 310|SP 150|SP 123|SP 270', case=False, regex=True, na=False)
df = df[mask]
```
